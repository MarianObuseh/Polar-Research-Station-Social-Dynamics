```python
# changes in centrality for end winter
eigenvector_centrality = nx.eigenvector_centrality(E_Graph)
print("The Eigenvector Centrality value of end winter graph is", eigenvector_centrality)
print("\n")

betweenness_centrality = nx.betweenness_centrality(E_Graph, normalized = False)
print("The Normalized Betweenness Centrality value of end winter graph is", betweenness_centrality)
print("\n")

closeness_centrality = nx.closeness_centrality(E_Graph)
print("The Normalized Closeness Centrality value of end winter graph is", closeness_centrality)
```

```python
pos1 = nx.spring_layout(eigenvector_centrality)
plt.figure(figsize = (10, 10))
nx.draw_networkx(E_Graph, node_list = eigenvector_centrality.keys(), node_size = [v * 10000 for v in eigenvector_centrality.values()], 
                with_labels = True, pos = pos1)
plt.axis('off')
plt.show()
```

```python
pos2 = nx.spring_layout(betweenness_centrality)
plt.figure(figsize = (10, 10))
nx.draw_networkx(E_Graph, node_list = betweenness_centrality.keys(), node_size = [v * 20 for v in betweenness_centrality.values()], 
                with_labels = True, pos = pos2)
plt.axis('off')
plt.show()
```
